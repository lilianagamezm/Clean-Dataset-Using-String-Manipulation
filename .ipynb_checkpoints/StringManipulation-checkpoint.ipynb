```python
import pandas as pd
import re
import numpy as np
import os
import glob
import warnings
warnings.filterwarnings('ignore')
```

```python
#local directory
Path=os.listdir(os.getcwd())
```

```python
#file name
fileName="TabulatedData.csv"
```

```python
#importing file from directory and converting to dataframe
for x in Path:
    if os.path.basename(x) in fileName:
        data=pd.read_csv(fileName)       
```

```python
data
```

```python
#convert data to string data type
data=data.astype(str)
```

```python
#find the positions of blanks in words 
print("Positions", "Word")
count=0
for i in data.columns:
    for ii in data[i].tolist():
        j=re.search("\s",ii)
        try:
            
            print(j.span(),ii)

            count+=1
        except:
            AttributeError
print("total words with blank space=",count)
```

```python
def CleanDataSet(data):
    for tc in data.columns:
        #remove special characters
        data[tc]=data[tc].str.replace(r"[^a-zA-Z0-9]", "")
        #convert each word to title case
        data[tc]=data[tc].str.title()
        #strip blank space
        data[tc]=data[tc].str.strip()
    return data
CleanDataSet(data)
```

```python
#save to csv
NewFileName=fileName.split(".")[0]
data.to_csv(NewFileName+"_cleaned.csv",index=False)
```
